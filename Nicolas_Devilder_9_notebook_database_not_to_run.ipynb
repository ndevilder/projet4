```python
import duckdb
import pandas as pd 
import os

import warnings
warnings.filterwarnings('ignore')
# Connexion à la base de données DuckDB
con = duckdb.connect('Projet4.db')
```

```python
# Définir le répertoire contenant les fichiers CSV
table_names=[]
directory_path ='ressource'
# Lire tous les fichiers CSV du répertoire et créer des tables dans DuckDB
for filename in os.listdir(directory_path):
    if filename.endswith(".csv"):
        file_path = os.path.join(directory_path, filename)
        table_name = filename.split('.')[0]  # Utiliser le nom du fichier sans l'extension comme nom de table
        table_names.append(table_name)
        try:
            df = pd.read_csv(file_path, encoding='utf-8')  # Spécifier l'encodage directement
            # Créer une table dans DuckDB
            con.execute(f"CREATE TABLE {table_name} AS SELECT * FROM df")
            print(f"Table {table_name} créée avec succès à partir du fichier {filename}")
        except Exception as e:
            print(f"Erreur lors de la lecture du fichier {filename}: {e}")

# Afficher les noms des tables créées
tables = con.execute("SHOW TABLES").fetchall()
print("Tables créées :", tables)
```
